```python
from utils import *
from preprocessing import *
from LDA import lda_analysis
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# Load data
df_list = read_df_list('combat_df_list_imms_1_full.pkl')
df_list_added_refuse = [[add_refuse_rate(df) for df in ls] for ls in df_list]
print(df_list_added_refuse[0])
```

```python
df_refuse_list = [[df for df in df_list if df['refuse'].values[0][0] > 0.1]for df_list in df_list_added_refuse  ]
df_acc_list = [[df for df in df_list if df['refuse'].values[0][0] <= 0.1]for df_list in df_list_added_refuse  ]
```

```python
docs_refuse = [' '.join([' '.join([row['text'] for ind, row in df.iterrows()]) for df in df_list]) for df_list in df_refuse_list]
docs_refuse = [str for str in docs_refuse if str != '']

docs_acc = [' '.join([' '.join([row['text'] for ind, row in df.iterrows()]) for df in df_list]) for df_list in df_acc_list]
docs_acc = [str for str in docs_acc if str != '']
```

```python
print(docs_refuse[1:5])
print(len(docs_refuse))

print(docs_acc[1])
print(len(docs_acc))
```

```python
n_topics_list = list(range(3, 21, 2))
results = lda_analysis(n_topics_list, TfidfVectorizer(), docs_refuse)

with open('lda_results_tfidf.pkl', 'wb') as f:
    pickle.dump(results, f)
print("Results have been saved to lda_results_tfidf.pkl")
```

```python
n_topics_list = list(range(2, 30, 1))
stop_words = ['gt', 'people']

results = lda_analysis(n_topics_list, TfidfVectorizer(), docs_refuse, stop_words=stop_words)
with open('lda_results_tfidf_ref.pkl', 'wb') as f:
    pickle.dump(results, f)
print("Results have been saved to lda_results_tfidf_ref.pkl")

results = lda_analysis(n_topics_list, CountVectorizer(), docs_refuse,filtermatrix=filter_matrix_c,stop_words=stop_words)
with open('lda_results_count_ref.pkl', 'wb') as f:
    pickle.dump(results, f)
print("Results have been saved to lda_results_count_ref.pkl")

results = lda_analysis(n_topics_list, CountVectorizer(), docs_acc, filtermatrix=filter_matrix_c, stop_words=stop_words)
with open('lda_results_count_acc.pkl', 'wb') as f:
    pickle.dump(results, f)
print("Results have been saved to lda_results_count_acc.pkl")

results = lda_analysis(n_topics_list, TfidfVectorizer(), docs_acc, stop_words=stop_words)
with open('lda_results_tfidf_acc.pkl', 'wb') as f:
    pickle.dump(results, f)
print("Results have been saved to lda_results_tfidf_acc.pkl")
```
