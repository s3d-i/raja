```python
# a function that runs ollama several times and record the results for a single piece of text

def run_ollama(text, top_k=30, top_p=0.8, n=5, sig='--IMMITATION--'):
    results = []
    for i in range(n):
        system_prompt = 'output format:\n{"'+f"{sig}"+'": your_immitation_texts}\n\nYour immitation should preserve the Speaker\'s semantic meaning and emotions sentence by sentence.\n'
        input_text = 'Speaker\'s Text:' + f"{text}"
        a = generate_answer_with_ollama(system_prompt=system_prompt, input_text=input_text, max_tokens=min(len(text), 30), top_k=top_k, top_p=top_p)
        results.append(a)
    return results

# a function that iterates over a list that contains lists containing dataframes, and for each dataframe, it runs the function above and save the results to a new column

def run_ollama_on_df_list(df_list, top_k=30, top_p=0.8, n=5, sig='--IMMITATION--'):
    for ls in df_list:
        for df in ls:
            df.loc[:, 'sentence_immitation'] = df['text'].apply(lambda x: run_ollama(x, top_k=top_k, top_p=top_p, n=n, sig=sig))
    return df_list

# save to pickle
def save_df_list(df_list, filename):
    with open(filename,df_list) as f:
        pickle.dump(df_list, f)
        
```
